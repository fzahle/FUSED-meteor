```python
from MeteorClient import MeteorClient

client = MeteorClient('ws://127.0.0.1:3000/websocket')
client.connect()
```

```python
def callback_function(error, result):
    if error:
        print(error)
        return

    print(result)

client.call('addItem', 'hello from python', callback_function)
```
